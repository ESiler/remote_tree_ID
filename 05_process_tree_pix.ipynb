```python
#load libraries
import os
from PIL import Image
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
## load data
df = pd.read_csv('../data/campustreedata.csv', index_col=None)

#remove trees with 0 lat or lon (indicating no location is recorded)
df = df[(df.LAT)!=0]
df = df[(df.LONG)!=0]
```

```python
#load image as np array
def load_image_to_flat_np(FID, filepath = "../Tree_Images/"):
    fp = filepath  + "FID_" + str(FID) + ".png"
    im = Image.open(fp)
    im_rgb = im.convert('RGB')
    npim = np.array(im_rgb).reshape((-1))
    return npim

def get_image_arrays(df=df):
    X_data = []
    fids = []
    for i, row in df.iterrows():
        fid = row['FID']
        npim = load_image_to_flat_np(fid)
        X_data.append(npim)
        fids.append(fid)
    return (np.array(X_data), fids)
```

```python
## X_data is the numeric information from all the individual tree images
X_data, fids = get_image_arrays()
```

```python
## X_data PCA
pca = PCA(n_components=100)
X_pca = pca.fit_transform(X_data)
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
## Save PCs for each tree image 
X_pca_df = pd.DataFrame(X_pca, index=fids)
#X_pca_df.to_csv("../data/single_tree_pcs.csv")
```
