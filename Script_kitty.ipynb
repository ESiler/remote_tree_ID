```python
#libraries
#!pip install folium geopy requests
import folium
from geopy.geocoders import Nominatim
```

```python
#input address, output GPS coordinates?
geolocator = Nominatim(user_agent="Ellie")

loc = geolocator.geocode("104 W 17th Ave, Spokane, WA, 99203")
```

```python
# input GPS coordinates, output map display?
lat, long = loc[1][0],loc[1][1]
mp = folium.Map(location=[lat, long], zoom_start=22, tiles='Esri.WorldImagery')

folium.Marker(
    location=[lat, long],
    tooltip="Click me!",
    popup=("Hi!!!!!!! This is Ellie! What's up????", lat, long),
    icon=folium.Icon(color="pink"),
).add_to(mp)

xoffset = .0005
yoffset = xoffset*1.3

folium.Rectangle(
    bounds=[[lat-xoffset, long-yoffset], [lat+xoffset, long+yoffset]],
    color='pink'
).add_to(mp)

mp
```
